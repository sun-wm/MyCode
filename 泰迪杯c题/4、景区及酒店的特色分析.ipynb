# 一、景区特色分析

```python
content = pd.read_excel(r"./附件1/景区评论.xlsx",engine='openpyxl')  
content.head()
```

```python
#使用停用词
def stopwordslist(filepath):
    stopwords = [line.strip() for line in open(filepath, 'r', encoding='utf-8').readlines()]
    return stopwords
```

```python
scenic_name = list(content.groupby('景区名称'))
names = []
for name in scenic_name:
    name = name[0]
    names.append(name)
names
```

```python
A01 = content.loc[content['景区名称']=='A01']
```

```python
def scenic_spot_participle(scenic_spot_name):
    stopwords = stopwordslist('stopwords.txt')
    words = jieba.lcut(' '.join(scenic_spot_name['评论内容'].values))
    for i in tqdm(words):
            if i in stopwords:
                words.remove(i)
    words = ' '.join(words)
    return words
```

```python
def words_name(words_name):
    vectorizer=CountVectorizer()         #该类会将文本中的词语转换为词频矩阵，矩阵元素a[i][j] 表示j词在i类文本下的词频  
    transformer=TfidfTransformer()       #该类会统计每个词语的tf-idf权值  
    X=vectorizer.fit_transform([words_name])   #将文本转为词频矩阵
    tfidf=transformer.fit_transform(X)   #计算tf-idf，  
    word=vectorizer.get_feature_names()  #获取词袋模型中的所有词语   
    weight=tfidf.toarray()               #将tf-idf矩阵抽取出来，元素a[i][j]表示j词在i类文本中的tf-idf权重  

    spec = []

    for  w in  weight:
            loc = np.argsort(-w)
            for i in range(50):
                print(u'-{}: {} {}'.format(str(i + 1), word[loc[i]], w[loc[i]]))
                spec.append((word[loc[i]], str(w[loc[i]])))
    return spec
```

```python
words_A01 = scenic_spot_participle(A01)
```

```python
words_A01_spec = words_name(words_A01)
```

```python
data = words_A01_spec


(
    WordCloud()
    .add(series_name="A01景区特色", data_pair=data, word_size_range=[6, 66])
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title="A01景区特色", title_textstyle_opts=opts.TextStyleOpts(font_size=23)
        ),
        tooltip_opts=opts.TooltipOpts(is_show=True),
    )
    .render_notebook()
)
```
