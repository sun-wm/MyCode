```python
import pandas as pd
from pandas import DataFrame
import numpy as np
import matplotlib.pyplot as plt
import jieba
import re
from collections import Counter
import csv
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
content = pd.read_excel(r"./附件1/景区评论.xlsx",engine='openpyxl')    
```

```python
# data = content.loc[content['景区名称']=='A01']
# data
```

```python
def filter_punc(sentence):
    sentence = re.sub("[\s+\.\!\/_,$%^*(+\"\'“”《》?“]+|[+——！，。？、~@#￥%……&*（）：]+", "", sentence)
    return sentence
```

```python

#使用停用词
def stopwordslist(filepath):
    stopwords = [line.strip() for line in open(filepath, 'r', encoding='utf-8').readlines()]
    return stopwords
```

```python
#分词
def participle(name,is_filter=True):
    all_words = []
    sentence = []
    data = content.loc[content['景区名称']==name]

    for idx,line in enumerate(data['评论内容']):
        if is_filter:         
            line = filter_punc(line)
       
        words = jieba.lcut(line)
        if len(words)>0:
            all_words = all_words+words
            sentence.append(words)
            
    print('{0} 包含 {1} 行, {2} 个词.'.format(name, idx+1, len(all_words)))
    dictions = {}
    cnt = Counter(all_words)
    
 
   
    for word,freq in cnt.items():
        dictions[word] = freq
    return sentence,dictions
    
```

```python
names = []
stopwords = stopwordslist('stopwords.txt')
for i in list(content.groupby('景区名称')):
    names.append(i[0])
for name in names:
    words = []
    sentence,dictions = participle(name,True)
    st = sorted(dictions.items(), key = lambda kv:(kv[1], kv[0]),reverse=True)
    # print(st)
    for i in st:
            if i[0] not in stopwords:
                words.append(i)
    with open(r'./热门词'+'/'+'景区'+'/'+name+'.csv','w',encoding='utf-8',newline='') as f:
        writer = csv.writer(f)
        writer.writerow(['评论热词','热度'])

        writer.writerows(words[:20])
```
